# Visualizations in Python on Sales Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Compare Sales by region for 2016 with 2015 using bar chart


```python
regional_sales=sales.groupby("Region")["Sales2015","Sales2016"].sum()
```

```python
regional_sales
```

```python
regional_sales.plot(kind="bar")
plt.ylabel('Sales')
plt.title("Sales by region for 2016 with 2015")
plt.show()
```

# 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Char

```python
regional_sales=sales.groupby("Region")["Sales2016"].sum()
```

```python
regional_sales
```

```python
plt.pie(regional_sales,autopct="%1.0f%%",labels=["Central","East","West"],shadow=True,colors=["b","g","r"],explode=[0.0,0.1,0.0])
plt.title("Contributing factors to the sales for each region in 2016")
plt.show()
```

# 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers

```python
sales_region_tier = sales.groupby(["Region","Tier"])['Sales2015','Sales2016'].sum()
sales_region_tier
```

```python
sales_region_tier.plot(kind="barh",stacked=True,figsize=(8,8))
plt.xlabel("Sales")
plt.ylabel("Region/Tier")
plt.title("Total sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

# 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported 

dataset. The Quarters are based on months and defined as -

• Jan - Mar : Q1

• Apr - Jun : Q2

• Jul - Sep : Q3

• Oct - Dec : Q4

```python
month =sales["Month"]
```

```python
quarter = []

for i in month:
    if i in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif i in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif i in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")    

        
    
        
        

```

```python
sales["quarter"]=quarter
```

```python
sales
```

```python
qtr_sales = sales.groupby("quarter")["Sales2015","Sales2016"].sum()
```

```python
qtr_sales
```

```python
qtr_sales.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

# 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart

```python
sales.head()
```

```python
qtr_pivot = sales.pivot_table(index='quarter',columns='Tier',values='Sales2016')
```

```python
qtr_pivot
```

```python
# Q1
plt.pie(x=qtr_pivot.loc["Q1",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['r', 'g', 'b'],shadow=True)
```

```python
#Q2
plt.pie(x=qtr_pivot.loc["Q2",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['r', 'g', 'b'])
plt.show()
```

```python
#Q3
plt.pie(x=abs(qtr_pivot.loc["Q3",:]),autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
#Q4
plt.pie(x=qtr_pivot.loc["Q4",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```
